```python
# First quartile (Q1)
Q1 = np.percentile(df["bmi"], 25, interpolation = 'midpoint')
  
# Third quartile (Q3)
Q3 = np.percentile(df["bmi"], 75, interpolation = 'midpoint')
  
# Interquaritle range (IQR)
IQR = Q3 - Q1
  
print(IQR)
```

```python
df["bmi"] = np.where(df["bmi"] <= 47.265, df["bmi"],47.265)
```

```python
def outlier_remover(data,column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3-Q1
  low = Q1-1.5*IQR
  high = Q3+1.5*IQR
  data[column] = np.where(data[column] <= high, data[column],high)
  data[column] = np.where(data[column] >= low, data[column],low)
  return data[column]

```

```python
df["charges"] = outlier_remover(df,"charges")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
scalar1=StandardScaler()
y_train=scalar1.fit_transform(pd.DataFrame(y_train,columns=["charges"]))
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost
```

```python
NUMERICAL = ['age',  'bmi']
CATEGORICAL = ["sex","smoker","region"]
numerical_pipe = Pipeline([
    ('scaler', StandardScaler())
])

categorical_pipe = Pipeline([
    ('encoder', OneHotEncoder())
])

preprocessors = ColumnTransformer(transformers=[
    ('num', numerical_pipe, NUMERICAL),
    ('cat', categorical_pipe, CATEGORICAL)
])

pipeline_lr = Pipeline([
    ('preprocessors', preprocessors),
    ('model', LinearRegression())
])

pipeline_rr=Pipeline([
    ('preproc', preprocessors),
    ('classifier', Ridge())])

pipeline_llr=Pipeline([
    ('preproc', preprocessors),
    ('classifier', Lasso())])

pipeline_rf=Pipeline([
    ('preproc', preprocessors),
    ('classifier', RandomForestRegressor())])

pipeline_xg=Pipeline([
    ('preproc', preprocessors),
    ('classifier', xgboost.XGBRegressor())])

```

```python
pipelines = [pipeline_lr, pipeline_rr, pipeline_llr,pipeline_rf,pipeline_xg]
```

```python
# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0: 'Linear Regression', 1: 'ridge regression', 2: 'lassoregression',3:'random forest',4:'xg boost'}

# Fit the pipelines
for pipe in pipelines:
	pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
for i,model in enumerate(pipelines):
    y_pred=model.predict(X_test)
    
    print("{} Mean Absolute  Error of : ".format(pipe_dict[i]))
    print("{0:.4f}".format(mean_absolute_error(y_test,scalar1.inverse_transform(y_pred.reshape(-1,1)))))
    print("{} Mean Squared Error of :".format(pipe_dict[i]))
    print("{0:.4f}".format(mean_squared_error(y_test,scalar1.inverse_transform(y_pred.reshape(-1,1)))))
    print("{} Root Mean Squared Error of :".format(pipe_dict[i]))
    print("{0:.4f}".format(np.sqrt(mean_squared_error(y_test, scalar1.inverse_transform(y_pred.reshape(-1,1))))))
```
